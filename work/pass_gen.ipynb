```python
import secrets
import string
import re

#kazu：パスワードの数を入力
kazu = 100
#パスワード生成関数。()内に桁数入れる
def pass_gen(size=12):
    chars = string.ascii_letters + string.digits
    chars = re.sub('[lIO01]', '', chars)
    return ''.join(secrets.choice(chars) for x in range(size))

for i in range(kazu):
  z=pass_gen(16)
  print(z)
```
